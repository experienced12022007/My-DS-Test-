## Q5

```python
dp['total']=dp["runs_of_bat"]+dp["extras"]
```

```python
total=dp.groupby("match_no")['total']
data=pd.DataFrame(data=total,columns=["match_no","Total Score"])
data['Total Score']=data['Total Score'].sum()
```

```python
sas.lineplot(data=data,x="match_no",y="Total Score",color="blue")
```

## Q6

```python
results=dp.groupby("bowler").agg({
    'over':'sum',
    "wicket_type":"notna",
})
dp["impact_score"]=dp["wickets"]*(dp["total_balls"]+1)/dp["economy"]
```

```python
sas.scatterplot(data=results,x="economy",y="impact_score")
```
